```python
import os, requests, bs4 
```

```python
url ='https://www.biqukan.com/50_50096/'

html = requests.get(url)
html.encoding = 'gbk'
obj = bs4.BeautifulSoup(html.text,'lxml')
title = obj.find('meta',property="og:title")
author = obj.find('meta',property="og:novel:author")
body = obj.find('meta',property="og:description")
print('標題為:',title['content'])
print('作者為:',author['content'])
print('內容為:',body['content'].strip())
```

```python
book =obj.find('div','listmain')
tt = book.find_all('dt')
print(tt[1].text)
print()

sto =book.find_all('dd')
sto = sto[12:]
for ch in sto:
    print (ch.text)
```

```python
chap=[]
abb = 'https://www.biqukan.com'
for ab in sto:
    ab_url = ab.a['href']
    chap.append(abb + ab_url)

    
for c in  chap:
    print(c)
```

```python
url = 'https://www.biqukan.com/50_50096/18520412.html'
html =requests.get(url)
html.encoding = 'gbk'
obj = bs4.BeautifulSoup(html.text,'lxml')
title = obj.find('h1')
story = ''+title.text + '\n'
print(title.text)
print()
con = obj.find('div',id='content')
print('區塊數量:', len(con))

for cc in con:
    if type(cc) == bs4.element.NavigableString:
        txt = cc.strip()
        if type(txt) == str and txt != '' :
            txt = cc.strip()
            txt = txt.replace('&1t;/p>', '')
            story = story + txt + '\n\n'
        
        
```

```python
fn = 'title.text'
with open( fn , 'w' , encoding='utf-8') as obj:
    obj.write(story)
print('%s 儲存成功'%title.text)
```
