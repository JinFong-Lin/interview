```python
import requests
from pprint import pprint
import csv
import random
import time
import pandas as pd
from urllib.parse import unquote
import matplotlib.pyplot as plt
from pylab import mpl
```

```python
url = 'http://www.lovewzly.com/api/user/pc/list/search?'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) \
            AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.117 Safari/537.36'}

form_data ={'gender':'2' , 'mary':'1','page':'1'}

datahtml = requests.get(url,params=form_data , headers=headers)
```

```python
data =datahtml.json()
print(type(data))
print(len(data))
pprint(data)
```

```python
print(data['data']['list'][0])
print('--'*20)
print('暱稱:',data['data']['list'][0]['username'])
print('USERID:',data['data']['list'][0]['userid'])
print('身高:',data['data']['list'][0]['height'])
print('薪資:',data['data']['list'][0]['salary'])
print('城市:',data['data']['list'][0]['city'])
print('學歷',data['data']['list'][0]['education'])
print('座右銘:',data['data']['list'][0]['monolog'])
print('出生年:',data['data']['list'][0]['birthdayyear'])
```

```python
fields = ['暱稱','出生年','學歷','身高','城市']
fn='out123.csv'
#abc = data['data']['list']

with open(fn,'w',newline='',encoding='utf-8') as csvfile:
    csvWriter = csv.writer(csvfile)
    csvWriter.writerow(fields)
    for d in data:
        d_list = [d['username'].encode('utf-8'),
                  d['birthdayyear'],
                  d['education'].encode('utf-8'),
                  d['height'],
                  d['city'].encode('utf-8')]
        csvWriter.writerow(d_list)
        print('暱稱為',d['username'])
        print('身高為',d['height'])
        print('-'*80)
```

```python
def get_data(page):
    form_data['page']=page
    print('正在抓取%s頁資料' %(page+1))
    datahtml =requests.get(url , params=form_data , headers = headers)
    datas = datahtml.json()
    data = datas['data']['list']
    for d in data: 
        d_list = [d['username'].encode('utf-8'),
                  d['birthdayyear'],
                  d['education'].encode('utf-8'),
                  d['height'],
                  d['city'].encode('utf-8')]
        csvWriter.writerow(d_list)
        print('暱稱為',d['username'])
        

with open(fn,'w',newline='',encoding='utf-8') as csvfile:
    csvWriter = csv.writer(csvfile)
    csvWriter.writerow(fields)
    for p in range(3):
        get_data(p)
        print('-'*30)
        time.sleep(random.randint(3,10))
```

```python
df = pd.read_csv('C:/Users/gg246/out123.csv',index_col=0)
edu = df['學歷']
eduction=[]
for txt in edu:
    txt =txt[1:].replace('\\x','%')
    txt = txt.replace("'",'')
    t = unquote(txt, encoding='utf-8')
    eduction.append(t)
    
eduction_count = {wd:eduction.count(wd) for wd in set(eduction)}

edu_list=[]
edu_count = []

for e ,c in eduction_count.items():
    print(e,c)
    edu_list.append(e)
    edu_count.append(c)
    
mpl.rcParams["font.sans-serif"] = ['SimHei']
plt.pie(edu_count,labels=edu_list,explode=(0,0,0.2,0,0,0),autopct="%1.2f%%")
plt.show()
```

```python
height = df['身高']

h_index=[0,0,0,0,0]

for h in height:
    if int(h)<150:
        h_index[0]+=1
    if int(h)<160:
        h_index[1]+=1
    if int(h)<170:
        h_index[2]+=1
    if int(h)<180:
        h_index[3]+=1
    else:
        h_index[4]+=1
        
fields = ['<150','15x','16x','17x','>180']

plt.bar(fields,h_index,width=0.35)
plt.ylabel('high')
plt.xlabel('num')
plt.show()
```

```python
year = df['出生年']

agg_index =[0,0,0,0,0]
for y in year:
    if int(y) <1980:
        agg_index[0]+=1
    if int(y) <1985:
        agg_index[1]+=1
    if int(y) <1990:
        agg_index[2]+=1
    if int(y) <1995:
        agg_index[3]+=1
    else:
        agg_index[4]+=1
        
field = ['<40','35-40','30-35','25-30','<25']

plt.bar(field,agg_index,width=0.35)
plt.ylabel('year')
plt.xlabel('num')
plt.show()
```
