```python
import requests , bs4

url = 'https://www.dcard.tw/f'
html  = requests.get(url)
obj = bs4.BeautifulSoup(html.text,'lxml')
```

```python
title = obj.find_all('div',class_='PostList_entry_1rq5Lf')
print(title[0].h3.text)
print(title[1].h3.text)
```

```python
header = title[2].find('span')
print('貼文論壇:',header.text)
school = title[2].find('span',class_='PostAuthor_root_3vAJfe')
print('學校為:',school.text)
time = title[2].find('span',class_='Header__PublishedDateTime-xvcbwe-3')
print('貼文時間為:',time.text)
print('標題為',title[2].h3.text)
body = title[2].find('div',class_='PostEntry_excerpt_2eHlNn')
print('內文為:',title[2].text)
good = title[2].find('div',class_="ActionBar__LikeCount-pwz3il-1")
print('按讚樹為:',good.text)
back = title[2].find('span',class_='ActionBar__CommentWrapper-pwz3il-5')
print('回復為:',back.text)
link = title[2].find('a')['href']
print('連結為:',link)
```
