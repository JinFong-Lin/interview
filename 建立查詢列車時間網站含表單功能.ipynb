```python
import bs4 , requests , json
```

```python
url = 'http://www.thsrc.com.tw/tw/TimeTable/SearchResult'
html = requests.get(url)
obj =bs4.BeautifulSoup(html.text , 'lxml')
```

```python
start = obj.find('select', id = 'StartStation').find_all('option')
for st in start:
    if st['value']:
        print(st.text.strip() , ':' , st['value'])
```
