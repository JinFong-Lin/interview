```python
import requests,bs4
import csv
```

```python
url ='http://www.taiwanrate.com/'
htmlfiles = requests.get(url)
htmlfiles.encoding = 'utf-8'
print('HTML編碼方式:',htmlfiles.encoding)
#print('列印網頁內容:',htmlfiles.text)
```

```python
obj = bs4.BeautifulSoup(htmlfiles.text,'lxml')
```

```python
table = obj.find_all('table')
```

```python
body = table[4].find('tr').find('tr').find('td')
```

```python
print(body.text,end=' ')
```

```python
tete = table[4].find('tr').find_all('a','bodytablehead')
```

```python
for he in tete:
    print(he.text,end='')
```

```python
rate = table[4].find('tr','bodytabletr1')
print(rate.text)
while rate.find_next_sibling('tr'):
    rate = rate.find_next_sibling('tr')
    print(rate.text)
```

```python
fn = 'out_13_4.csv'
tablelist = []
headlist = []
ratelist = []


for he in tete:
    headlist.append(he.text)

for head in rate :
    headlist.append(head.text)

tablelist.append(headlist)

rated = table[4].find('tr','bodytabletr1')
for row in rated:
    ratelist.append(row.text)
tablelist.append(ratelist)

while rated.find_next_sibling('tr'):
    rated = rated.find_next_sibling('tr')
    ratelist = []
    for row in rated:
        ratelist.append(row.text)
    tablelist.append(ratelist)
    
    
    
    
with open(fn,'w',newline = '') as csvFile:
    csvWriter = csv.writer(csvFile)
    for row in tablelist:
        csvWriter.writerow(row)
```

```python
html = 'https://www.moneydj.com/funddj/ya/yp401000.djhtm'
abc = requests.get(html)
home = bs4.BeautifulSoup(abc.text,'lxml')
```

```python
headlist = []
money = []


table = home.find('table' , id='oMainTable')
obj = table.find('tr',id='oScrollMenu')
body = obj.find_all('th')
for ab in body:
    headlist.append(ab.text.strip())

print(headlist)
```

```python
monwey = []
money.append(headlist)
omg = table.find('tbody')
apex = omg.find_all('tr')
for table in apex:
    row = table.text.strip()
    txt = row.split('\n')
    money.append(txt)
    
print(money)
```

```python
with open('out_4_3.csv' , 'w' , newline= '') as csvFile:
    csvWriter = csv.writer(csvFile)
    for row in money:
        csvWriter.writerow(row)
```
