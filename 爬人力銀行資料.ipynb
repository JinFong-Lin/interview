```python
import requests , bs4
import json
import time
import random
```

```python
urls = {'https://www.104.com.tw/jobs/search/?keyword=python&order=1&jobsource=2018indexpoc&ro=0',
      'https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=python&order=14&asc=0&page=2&mode=s&jobsource=2018indexpoc',
       'https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=python&order=14&asc=0&page=3&mode=s&jobsource=2018indexpoc'
      }
url =('https://www.104.com.tw/jobs/search/?keyword=python&order=1&jobsource=2018indexpoc&ro=0')
html = requests.get(url)
obj = bs4.BeautifulSoup(html.text,'lxml')
```

```python
job = obj.find_all('article',class_='js-job-item')
job_list = []
for a in job:
    name = a.get('data-cust-name')
    print('公司名稱:',name)
    names =a.get('data-job-name')
    print('職位名稱:',names)
    print('-'*20)
    d = [('公司名稱',name) , ('職位名稱',names)]
    j_dict = dict(d)
    job_list.append(j_dict)
    
myjob = {'Job':job_list}
fn ='ch25_2.json'
with open(fn,'w') as fnObj:
    json.dump(myjob,fnObj,indent=2,ensure_ascii=False)
```

```python
def job_list(url):
    html = requests.get(url)
    obj = bs4.BeautifulSoup(html.text,'lxml')
    jobs = obj.find_all('article',class_='js-job-item')
    for job in jobs:
        print ('公司名稱:', job.get('data-cust-name'))
        print('職位名稱:',job.get('data-job-name'))

        
for url in urls:
    job_list(url)
    print('---'*70)
    time.sleep(random.randint(1,5))
```

```python
a ='https://www.104.com.tw/jobs/search/?ro=0&kwop=7&keyword=python&order=14&asc=0&page='
b='&mode=s&jobsource=2018indexpoc'
page_total = 5
for page in range(page_total):
    url = a+str(page_total)+b
    job_list(url)
    print('--'*20)
    time.sleep(random.randint(3,5))
```
