```python
from pygal.maps.world import COUNTRIES
```

```python
for countryCode in sorted(COUNTRIES.keys()):
    print ('國家代碼:',countryCode,
          '國家名稱 = ',COUNTRIES[countryCode])
```

```python
def getCountryCode(countryName):
    for dictCode,dictName in COUNTRIES.items():
        if dictName == countryName:
            return dictCode
    return None
```

```python
dictData={}
for getdata in getDatas:
    if getdata['Year'] == '2000':
        countryName = getdata['Country Name']
        countryCode = getCountryCode(countryName)
        population =int(float(getdata['Numbers']))
        if countryCode != None:
            dictData[countryCode] = population
```

```python
dict1 , dict2 ={},{}
```

```python
for code, population in dictData.items():
    if population > 100000000:
        dict1[code] = population
    else:
        dict2[code] = population
```

```python
import pygal.maps.world
```

```python
worldMap = pygal.maps.world.World()
worldMap.title = '人口總數'
worldMap.add('voer 1E',dict1)
worldMap.add('other',dict2)
worldMap.render_to_file('out1_17.svg')
```
