```python
import bs4 ,requests ,os
```

```python
url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
htmlfile = requests.get(url)
html = bs4.BeautifulSoup(htmlfile.text,'lxml')
```

```python
num = 0
items = html.find_all('div','release_info')
for item in items:
    cName = item.find('div','release_movie_name').a.text.strip()
    rTime = item.find('div','release_movie_time')
    enName = item.find('div','en').a.text.strip()
    ea = item.find('div','leveltext').text.strip()
    book = item.find('div','release_text').text.strip()
    photo = item.find_previous_sibling('div','release_foto').a.img['src']
    viedo = item.find('div','release_btn color_btnbox').find_all('a')[1]
    if 'href' in viedo.attrs:
        viedo = viedo['href']
    else:
        viedo=''
    num +=1
    print('中文名稱:',cName)
    print('英文名稱:',enName)
    print('期待度:',ea)
    print('編號:',num)
    print('海報網址:',photo)
    print('摘要:',book)
    print('預告片:',viedo)
    print('-'*20)
```

```python
ptotos = []
pho = html.find_all('div','release_foto')

for item in pho:
    photo = item.a.img['src']
    ptotos.append(photo)

des = 'out_95'
if os.path.exists(des) ==False:
    os.mkdir(des)  #沒有則建立
    
print('搜尋到的圖片數量:',len(ptotos))

for pgo in ptotos:
    picture = requests.get(pgo)
    picture.raise_for_status()
    print('%s 圖片下載成功 ' %photo)
    
    pictfile = open(os.path.join(des,os.path.basename(pgo)),'wb')
    for disk in picture.iter_content(10240):
        pictfile.write(disk)
    pictfile.close()
```
